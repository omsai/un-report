```python
# Sort data with sort_values()
```

```python
(
    gapminder
    .query('year == 2007')
    .groupby('continent')
    .agg({'lifeExp' : 'mean'})
    .sort_values('lifeExp', ascending=False)
)
```

```python
# Changing the shape of data
```

```python
(
    gapminder
    .filter(['continent', 'country', 'lifeExp', 'year'])
    .pivot(columns='year',
           values='lifeExp',
           index=['continent', 'country'])
)
```

```python
(
    gapminder
    .filter(['continent', 'country', 'lifeExp', 'year'])
    .pivot(columns='year',
           values='lifeExp',
           index=['continent', 'country'])
    .reset_index()
    .melt(id_vars=['continent', 'country'],
          value_name='lifeExp')
)
```
